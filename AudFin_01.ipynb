# Cálculo dos Saldos Finais ao Término de Cada Mês

---

A base de dados possui o saldo inicial de cada conta para cada órgão mas não traz os saldos ao término de cada més. Vamos calcular esses valores:

```python
# audfin <- audfin %>%
#             mutate(janSdFin = ifelse(grepl("^1|[(-)]", cod_conta),  sd_inic + janD - janC, sd_inic + janC - janD),
#                    fevSdFin = ifelse(grepl("^1|[(-)]", cod_conta), janSdFin + fevD - fevC, janSdFin + janC - janD),
#                    marSdFin = ifelse(grepl("^1|[(-)]", cod_conta), fevSdFin + marD - marC, fevSdFin + fevC - fevD))

(audfin
    .assign(
        ajuste_ou_diminuicao = audfin['cod_conta'].str.contains("\(-\)")*2-1,
        jan_sd_fin=lambda x: (x['sd_inic']     + x['ajuste_ou_diminuicao']*(x['janC'] - x['janD'])),
        fev_sd_fin=lambda x: (x['jan_sd_fin']  + x['ajuste_ou_diminuicao']*(x['fevC'] - x['fevD'])),
        mar_sd_fin=lambda x: (x['fev_sd_fin']  + x['ajuste_ou_diminuicao']*(x['marC'] - x['marD'])))
    .drop('ajuste_ou_diminuicao', axis = 1)
    .head())
```
